```python
import matplotlib.pyplot as plt
```

```python
x = ["w1", "w2", "w3"]
y1 = [71, 70, 55]
y2 = [410, 511, 661]
```

```python
fig, ax1 = plt.subplots(figsize=(16, 8))

color = "#A95AA1"
ax1.plot(x, y1, "-o", markersize=32, color=color)
ax1.set_xlabel('week')
ax1.set_ylabel('cloners', color=color)
ax1.set_ylim([0, max(y1)*1.1])
ax1.tick_params('y', colors=color)

color = "#0F2080"
ax2 = ax1.twinx()
ax2.plot(x, y2, "-o", markersize=32, color=color)
ax2.set_ylabel('visitors', color=color)
ax2.set_ylim([0, max(y2)*1.1])
ax2.tick_params('y', colors=color)

plt.grid()
plt.title("# unique visitors, cloners")
```
