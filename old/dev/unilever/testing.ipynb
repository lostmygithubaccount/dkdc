```python
import numpy as np
import pandas as pd
```

```python
from azureml.core import Workspace, Dataset, Experiment
from azureml.train.automl import AutoMLConfig
```

```python
ws = Workspace.from_config()
```

```python
dset = Dataset.get_by_name(ws, 'indaiatuba')
```

```python
df = dset.to_pandas_dataframe()
```

```python
df.head()
```

```python
df.info()
```

```python
df['ts']
```

```python
X = df.copy()
y = np.array(df[list(df.columns[df.columns.str.contains('Moisture')])[0]])
```

```python
SHIFT = 20

y = np.array([0]*SHIFT + list(y)[:-SHIFT])
```

```python
automl_config = AutoMLConfig(task='regression',
                             primary_metrics='normalized_root_mean_squared_error',
                             iterations=10,
                             X=X.copy(),
                             Y=y.copy(),
                             n_cross_validations=5,
                             preprocess=False
                            )
```

```python
experiment_name = 'indaiatuba_automl_1'
experiment = Experiment(ws, experiment_name)
```

```python
run = experiment.submit(automl_config, show_output=True)
```
