```python
from azureml.core import Dataset, Workspace
```

```python
ws = Workspace.from_config()
```

```python
ws
```

```python
dstore = ws.get_default_datastore()
```

```python
dstore.upload('data', 'tabular-data')
```

```python
baseline = Dataset.Tabular.from_parquet_files(dstore.path('tabular-data/2019/01/data.parquet'))
```

```python
baseline.register(ws, 'baseline7')
```

```python
target = Dataset.Tabular.from_parquet_files(dstore.path('tabular-data/2019/*/data.parquet'))
```

```python
target = target.with_timestamp_columns('datetime')
```

```python
target.register(ws, 'target7')
```
