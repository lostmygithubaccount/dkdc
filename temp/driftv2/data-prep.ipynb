```python
import os

import numpy as np
import pandas as pd 

from azureml.core import Workspace, Dataset, Model

from azureml.opendatasets import NoaaIsdWeather

from datetime import datetime, timedelta
```

```python
ws = Workspace.from_config()
dstore = ws.get_default_datastore()
```

```python
start = datetime(2019, 1, 1)
end   = datetime(2019, 2, 1)
```

```python
isd = NoaaIsdWeather(start, end)
```

```python
df = isd.to_pandas_dataframe()
```

```python
df = df[df['stationName'].str.contains('FLORIDA', regex=True, na=False)]
```

```python
os.makedirs('my-data', exist_ok=True)
```

```python
df.to_csv('my-data/training.csv', index=False)
```

```python
dstore.upload('my-data', 'my-data')
```

```python
dset = Dataset.Tabular.from_delimited_files(dstore.path('my-data/training.csv'))
```

```python
dset.register(ws, 'my-dataset')
```
