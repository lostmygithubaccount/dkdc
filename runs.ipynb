```python
units = ['A', 'B', 'C', 'D']
```

```python
import os
os.mkdir('scripts')
```

```python
%%writefile scripts/run.py
print('hello world')
```

```python
from azureml.core import Workspace
ws = Workspace.from_config()
ws
```

```python
ct = ws.compute_targets['cpu-cluster']
ct
```

```python
from azureml.train.estimator import Estimator
from azureml.core.runconfig import MpiConfiguration

estimators = []

script_params = {
    '--data-folder': 'None'
}

for unit in units:
    estimators.append(Estimator(source_directory='scripts',
                      compute_target = ct,
                      entry_script='run.py',
                      script_params=script_params,
                      node_count=4,
                      process_count_per_node=1,
                      distributed_training=MpiConfiguration(),   
                      ))
```

```python
from azureml.core import Experiment
exp = Experiment(ws, 'my-exp')
```

```python
runs = [exp.submit(est) for est in estimators]
```

```python
from azureml.widgets import RunDetails
```

```python
RunDetails(runs[0]).show()
```
